# Trabalhando os dados

## Carregando o dataset

```python
import pandas as pd
import numpy as np
```

```python
ds = pd.read_csv('dataset/train.csv', header=0)
print(ds.shape)
ds.head(15)
```

```python
# ds.head(200).to_csv('dataset/train_slice.csv', index=False)
```

## Ordenando o dataset

```python
ds['id_num'] = ds['id'].str[4:].astype(int)
```

```python
ds = ds.sort_values(by=['p_num', 'id_num'], ascending=[True, True])
ds.head(15)
```

## Convertendo o time num inteiro crescente

```python
ds['minutes_since_start'] = (pd.to_datetime(ds['time'], format="%H:%M:%S").dt.hour * 60 + pd.to_datetime(ds['time'], format="%H:%M:%S").dt.minute)
ds.head(15)
```

```python
ds["time_circle"] = np.sin(2 * np.pi / 720 * ds['minutes_since_start']) + 2
ds.head(15)
```
