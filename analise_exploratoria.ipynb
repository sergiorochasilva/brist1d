# Análise Exploratória

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Carregando o dataset de treino

```python
ds = pd.read_csv('dataset/train.csv', header=0)
print(ds.shape)
ds.head(5)
```

## Verificando o percentual de nulos por coluna

```python
# Verificando a quantidade de nulos
print(len(ds))

nulos_por_coluna = ds.isna().sum()
print(nulos_por_coluna)

# Calcula o percentual de nulos por coluna
percentual_nulos = ds.isna().mean() * 100
print(percentual_nulos)
```

### Nível de Glicose

```python
# Filtrando as variáveis de nível de glicose
nulos_filtrado = percentual_nulos.filter(like='bg-')

# Plota o gráfico de barras
plt.figure(figsize=(20, 10))
nulos_filtrado.plot(kind='bar', color='skyblue')

plt.ylim(0, 100)

# Adiciona título e rótulos
plt.title('Percentual de Valores Nulos por Coluna', fontsize=14)
plt.ylabel('Percentual (%)', fontsize=12)
plt.xlabel('Colunas', fontsize=12)

# Exibe o gráfico
plt.show()
```

### Quantidade de Insulina
